```python
import pandas as pd
import matplotlib as mp
import matplotlib.pyplot as plt
import pymysql

db = pymysql.connect(host="localhost", user="rnv_importer", password="rnv_importer", database="rnv_big_data_mining")
cursor = db.cursor()
#linegroups, that are most often late (most frequently late)
cursor.execute("SELECT COUNT(*) as anz, `lines`.linegroup, MAX(TIMESTAMPDIFF(SECOND, api_plannedDeparture, api_realtimeDeparture)) AS latency FROM stops INNER JOIN `lines` ON `lines`.id = stops.api_line WHERE (api_plannedDeparture != api_realtimeDeparture) GROUP BY `lines`.linegroup ORDER BY anz DESC LIMIT 10")
latencyPerLine = cursor.fetchall()

# put in lists for chart
anz = []
linegroup = []
latency = []

for record in latencyPerLine:
    anz.append(record[0])
    linegroup.append(record[1])
    latency.append(record[2])

#linechart: how many times the is late
df = pd.DataFrame({'H채ufigkeit': anz}, index=linegroup)
ax = df.plot.bar(rot=0, color="blue")
#linechart: how late the line max is
df = pd.DataFrame({'Versp채tung': latency}, index=linegroup) 
# I tried:
#df['Versp채tung'] = pd.to_datetime(df["Versp채tung"], unit='s')
ax = df.plot.bar(rot=0, color="red")
```
